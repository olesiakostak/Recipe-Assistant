```python
def find_recipes_by_ingredient(ingredient_name: str):
    ingredient_name = ingredient_name.lower()
    return data[data['ingredients'].str.lower().str.contains(ingredient_name, na=False)]['title'].tolist()
```

```python
recipe_title = find_recipes_by_ingredient('chicken')
recipe_title
```

```python
def get_recipe_details(recipe_title: str):
    row = data[data["title"] == recipe_title]
    if row.empty: return f"Recipe {recipe_title} is not found"
    return {"ingredients": row["ingredients"].iloc[0],
            "NER": row["NER"].iloc[0],
            "recipe": row["directions"].iloc[0]}

get_recipe_details("Jewell Ball'S Chicken")
```
